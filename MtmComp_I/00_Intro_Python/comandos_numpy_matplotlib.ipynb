````python
import numpy as np

#Criação de vetores:

#Vetor com três entradas (ou uma matriz)
x = np.array([1,2,3])
print(f'Vetor x: {x}')

A = np.array([[1,2,3],[1,2,3]])
print(f'Matriz A:')
print(A)

print('*'*50)
#Podemos usar o comando .shape para descobrir a dimensão de matrizes e vetores:
A.shape
````

````python
# Com numpy, o operador * opera elemento a elemento. Por exemplo:

A = np.array([[1,2],[2,1]])
print(A)
B = np.array([[2,1],[2,1]])
print(B)

print('*'*50)
print('Produto de A por B, usando *')
print(A*B)

# Para fazer o produto de matrizes, usamos o .dot():
C = A.dot(B)

print('*'*50)
print('Produto de matrizes AxB:')
print(C)

# A transposta é obtida por .transpose()
# pode ser utilizado apenas .T também.
D = A.transpose()

print('*'*50)
print('Matriz transposta de A:')
print(D)

# O seguinte código faz o cálculo de A^t x A x B:
E = A.transpose().dot(A).dot(B)

print('*'*50)
print('cálculo de A^t x A x B:')
print(E)

# As operações de adição e subtração são feitas diretamente com + e -.
# As operações de divisão e potenciação, / e **, respectivamente, são efetuadas termo a termo.

# Podemos criar uma matriz apenas com 1 em todas entradas pelo np.ones((n_linhas, n_colunas))
print('*'*50)
F = np.ones((3,5))
print('Matriz com todos elementos iguais a 1 de ordem 3 por 5:')
print(F)

# O mesmo pode ser feito com np.zeros((n_linhas, n_colunas))
print('*'*50)
G = np.zeros((2,2))
print('Matriz com todos elementos iguais a 0 de ordem 2 por 2:')
print(G)
````

````python
# Podemos criar um vetor com uma sequência de numeros espaçados igualmente por np.arange().
# A sintaxe o np.arange é: np.arange(inicio, fim, intervalo)
z = np.arange(1, 50, 1)
print(z)

print('*'*50)
z = np.arange(1, 100, 0.5)
print(z)

print('*'*50)
# O comando np.linspace faz algo similar, porém indicamos a quantidade de subintervalos queremos dividir o intervalo
# A sintaxe o np.linspace é: np.linspace(inicio, fim, intervalo). O cálculo é feito por (fim - inicio)/intervalo
w = np.linspace(1, 50, 10)
print(w)

# Note a diferença nas saídas abaixo: o arange cria um vetor que começa em 1 e vai criando números até chegar em 100, com intervalor de 0.5
# Já no linspace, o intervalo [1,50] é dividido em 10 subintervalos, da seguinte maneira: (50-1)/10
````

````python
x = np.array([1, 4, 3, 4, -1, 6, 7, -5])

# Para identificar o segundo elemento de x: Usamos x[1] -> lembre que no python, a indexação  começa no 0.
print(x[1])

# Para identificar o primeiro elemento de x:
print(x[0])

# Para identificar o último elemento de x:
print(x[-1])

# Selecionar todos elementos do segundo até o último: x[1:]
print(x[1:])

# Selecionar todos elementos, do primeiro até o quarto: x[:3]
print(x[:3])

# Selecionar todos elementos entre o segundo e o sexto: x[1:5]
print(x[1:5])

# Para matrizes, temos:
# A[:,:]   toda a matriz
# A[i:j,k] os elementos das linhas i até j (exclusive) da k-ésima coluna
# A[i,j:k] os elementos da i-ésima linha das colunas j até k (exclusive)
# A[i,:]   a i-ésima linha da matriz
# A[:,j]   a j-ésima coluna da matriz
````

````python
# O numpy suporta diversas funções matemáticas intrinsicas. A raiz quadrada de um número pode ser calculada por:
# np.sqrt()

raiz = np.sqrt(100)
print(raiz)

print('*'*50)
# Ou podemos calcular o seno e cosseno de angulos (aqui, usaremos np.pi para o pi):

seno = np.sin(np.pi)
print('Seno de pi:')
print(seno)
cos  = np.cos(np.pi)

print('*'*50)
print('Cosseno de pi:')
print(cos)

# Note na saida que o seno fornece um numero absurdamente próximo de zero, mas não zero (temos aqui um erro de arredondamento)
````

Outras funções matemáticas com o numpy podem ser encontradas [aqui](https://numpy.org/doc/stable/reference/routines.math.html).

## Criação de gráficos:

Usaremos o módulo matplotlib.pyplot:

````python
import matplotlib.pyplot as plt

# Aqui, precisamos criar o domínio e a imagem em vetores:
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

# O gŕafico de uma função em função de uma variável tem uma sintaxe simples: basta usar plt.plot(eixo_x, eixo_y)
plt.plot(x,y)
````

````python
# Se quiseremos diferentes gráficos em uma mesma células, temos que usar o plt.show() abaixo de cada plot.
z = np.cos(x)

plt.plot(x,y)
plt.show()

plt.plot(x,z)
plt.show()
````

````python
# Mas podemos colocar tudo numa mesma figura, definindo o dominio e imagem dois a dois:
plt.plot(x,y,x,z)
plt.show()

# Ou podemos fazer separando os plots:
plt.plot(x,y)
plt.plot(x,z)
plt.show()

# Veja abaixo que os mesmos gráficos são criados.
````

````python
# Podemos adicionar o grid no gráfico, usando
plt.plot(x,y)
plt.grid()
plt.show()
````

````python
# E podemos mudar o tipo da linha e a cor com:
plt.plot(x,y, 'ro')
plt.plot(x,z, 'b--')
plt.grid()
plt.show()

# Aqui nesse comando, r é de red e o indica uma bola, b é azul e -- uma linha tracejada.
````

````python
# Podemos adicionar a legenda identificando cada gráfico por 'label' dentro do plot:
plt.plot(x,y, 'ro', label = 'y = sin(x)')
plt.plot(x,z, 'b--', label = 'y = cos(x)')
plt.grid()
plt.legend()
plt.show()
````

````python
# Construir y = sen(x)
import numpy as np

def f1(x):
  return np.sin(x)

def f2(x):
  return np.cos(x)
````

````python
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi, 1000)

plt.plot(x, f1(x), x, f2(x))
plt.grid()
````

````python
# y = x + sin(x)

def g(t):
  return t + np.sin(t)

x = np.linspace(-2*np.pi,2*np.pi, 1000)

plt.plot(x, g(x))
plt.grid()
````
