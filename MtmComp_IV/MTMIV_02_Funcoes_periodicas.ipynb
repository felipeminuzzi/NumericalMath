Uma função periódica é definida como aquela que apresenta repetição de seus valores a um intervalo definido de tempo, denominado período:

$$f(t) = f(t + T),$$

onde o período $T$ é o comprimento deste intervalo. O menor valor de $T$ que satisfaz a equação acima é chamado **período fundamental** da função.


**Extensões pares e ímpares**</u>

Funções cujo gráfico é <u>simétrico em relação ao eixo $y$</u>, isto é, que satisfazem

$$
f(-x) = f(x),
$$

são chamadas **funções pares**. Por exemplo:

<br>

<center><img src="https://github.com/guerinileonardo/figures/blob/main/fun_par.png?raw=true" width="300"></center>

<br>

Por outro lado, funções que satisfazem

$$
f(-x) = -f(x)
$$

são <u>simétricas em relação à origem</u>, e chamadas **funções ímpares**.

Por exemplo:

<center><img src="https://github.com/guerinileonardo/figures/blob/main/fun_impar.png?raw=true" width="300"></center>

**Obs:** Note que a função cosseno é uma função par e a função seno é uma função ímpar.
