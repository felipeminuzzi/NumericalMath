```python
"""
Script to test the finite difference derivatives.
"""

import os
os.chdir('/Users/felipeminuzzi/Library/Mobile Documents/com~apple~CloudDocs/Documents/NumericalMath/')

import numpy as np
import src.derivatives as der

%load_ext autoreload
%autoreload 2

diff = der.FirstOrder()
```

### First Derivative

```python
x        = np.empty([1, 50])
x[0, :]  = np.linspace(0, 2*np.pi, 50)
dx       = x[0, 1] - x[0, 0]

x2       = np.empty([1, 50])
x2[0, :] = np.linspace(-2, 2, 50)
dx2      = x2[0, 1] - x2[0, 0]

y  = np.sin(x)
y2 = x2**2 

y_prime_1 = diff.differentiateFirstFO(x, y, dx)
y_prime_2 = diff.differentiateFirstFO(x2, y2, dx2)
```

```python
from matplotlib import pyplot as plt
y_prime_exact = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x[0,:], y_prime_1[0,:], linestyle = ':', label='first order first derivative', linewidth=3)
ax.plot(x[0,:], y_prime_exact[0, :], label='exact', linewidth=1)

plt.xlim(0.0, 2.0*np.pi)
plt.ylim(-1.0, 1.0)
plt.tick_params(axis='both', which='major', labelsize=12)

plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

legend = ax.legend(loc='upper center', shadow = False)
```

```python
y_prime_exact = 2*x2

fig, ax = plt.subplots()
ax.plot(x2[0,:], y_prime_2[0,:], linestyle = ':', label='first order first derivative', linewidth=3)
ax.plot(x2[0,:], y_prime_exact[0, :], label='exact', linewidth=1)

plt.xlim(-2, 2)
plt.ylim(-4.0, 4.0)
plt.tick_params(axis='both', which='major', labelsize=12)

plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

legend = ax.legend(loc='upper center', shadow = False)
```
